```python
chunk1 = (.745/.255) / ( (1+(.745/.255))**2 )
chunk2 = .0011 * (1 + (.745/.255) )
a = (chunk1 - .0011) / chunk2
b = a * (.745/.255)
```

```python
mu = pymc.Beta(name='mu', alpha=a, beta=b, value=None)
mu.random()
```

```python
@pymc.deterministic(plot=False)
def modelled_p(mu=mu):
    #mus = np.zeros(n_count_data)
    #mus[:] = mu
    return mu
```

```python
#observation = pm.Poisson("obs", lambda_, value=count_data, observed=True)
xi = pymc.Binomial('xi', n=aprilD.AB[0], p=modelled_p, value=aprilD.H[0], observed=True)

#model = pm.Model([observation, lambda_1, lambda_2, tau])
model = pymc.Model([xi,mu])
```

```python
#mcmc = pm.MCMC(model)
#mcmc.sample(40000, 10000, 1)
mcmc = pymc.MCMC(model)
mcmc.sample(40000, 10000, 1)
```

```python
mu_samples = mcmc.trace('mu')[:]
print "posterior mean: " , mu_samples.mean()
sorted_ms = sorted(mu_samples)
print "posterior 95% CI, lower limit: " , sorted_ms[int(len(mu_samples)*.05)]
print "posterior 95% CI, upper limit: " , sorted_ms[int(len(mu_samples)*.95)]
```

```python
pymc.Matplot.plot(mcmc)
```
