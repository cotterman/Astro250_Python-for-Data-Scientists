```python
mydf = pd.read_csv(towrite_path) #the transform(np.mean) does not want to have a time series index
#create age_group variable
age_group_list = []
for row_num in xrange(mydf.shape[0]):
    #from text descriptions, 
        #it's clear that database lumped 0-yr olds with 1-yr olds
    if mydf.patient_age[row_num]==1:
        age_group="0_infant (0-1)"
    elif mydf.patient_age[row_num]<6:
        age_group="1_toddler/preschool (2-5)"
    elif mydf.patient_age[row_num]<12:
        age_group="2_school-age (6-11)"
    elif mydf.patient_age[row_num]<20:
        age_group="3_preteen/teen (12-19)"
    elif mydf.patient_age[row_num]<30:
        age_group="4_twenty-something (20-29)"
    elif mydf.patient_age[row_num]<40:
        age_group="5_thirty-something (30-39)"
    elif mydf.patient_age[row_num]>=40 and pd.notnull(mydf.patient_age[row_num]):
        age_group="6_forty plus (40+)"
    else:
        age_group=None

    age_group_list.append(age_group)

mydf['age_group']=age_group_list
print "Age groups \n"
print mydf.age_group.value_counts(normalize=True) , "\n"
print mydf.age_group.value_counts(normalize=False) , "\n"
```

```python
#(use mydf from above)
#average smile size by country and age 
means = mydf.groupby(['region','age_group']).transform(np.mean)
mean_by_c = pd.DataFrame(data=means, index=mydf.index)
#print mean_by_c
dfnow = pd.DataFrame(data=mydf[['region','age_group']], index=mydf.index)
toplot = mean_by_c.merge(dfnow, right_index=True, left_index=True)

dedup = toplot.drop_duplicates(['region','age_group'])
dedup = dedup[['region','smile_scale','age_group']]
dedup = dedup.dropna(axis=0) #had a bunch of NaN rows to get rid of 
#print "dedup here\n" , dedup #good
#print dedup['region']
dedup.index = dedup['region']
#dedup = dedup.drop_duplicates(['region','age_group'])
#print dedup
#print type(dedup)
#print dedup
dedup = dedup.sort_index(by=['region','age_group'], ascending=True)
print dedup[['age_group','smile_scale']]
#dedup.sort_index?
mycolors = ['b']*7+['r']*7+['g']*7
print mycolors
dedup.plot(kind='barh', stacked=False, alpha=0.5, figsize=(15,9), legend=False, color=mycolors)
#dedup.plot?

#how can I get the age_group labels incorporated into this graph??
#might need to rearrange data (as in 1way notebook bar graph) so that columns are region,
 #rows are age_group (and is the index), and smile_scale is in cells (from long to wide)
    #(see "reshaping with hierarchical indexing" in manual)
ytick_names1 = ["infant (0-1)","toddler/preschool (2-5)","school-age (6-11)",\
       "preteen/teen (12-19)","twenty-something (20-29)","thirty-something (30-39)", \
        "forty plus (40+)"]
newticks = ytick_names1*3

plt.yticks(xrange(21), newticks, rotation='horizontal', fontsize=12)
plt.ylabel("Region/age group")
plt.xlabel("Smile scale (0 - 1)")
plt.savefig("02_Smile_by_Age_and_region.pdf")

print "\nCell sizes (age group, region crosstabulation)\n"
a_r_counts = pd.crosstab(mydf.age_group, mydf.region)
print a_r_counts
```

```python
#create time series with index indicating monthly period
mydf.index=mydf['date_posted']
#print "time series: \n", mydf[:5] 

myts_monthly = mydf.to_period(freq='M') #convert to data with a monthly "period" index
#print "\nWith index indicating month:\n" , myts_monthly['profile_ID'].ix[:5]

#monthly averages in which each row represents 1 month
mydf_M_avg = mydf.resample('M', how='mean').to_period(freq='M')
#print myts_M_avg.columns
mydf_M_avg = mydf_M_avg.drop(['Unnamed: 0','profile_ID'], axis=1)
#myts_M_avg.drop('profile_ID', axis=1)
#mydf.resample?
#print "\nMonthly averages: " , myts_M_avg.ix[:]

#number of profiles posted each month 
myts_M_count = mydf['profile_ID'].resample('M', how='count').to_period(freq='M')
mydf_M_count = pd.DataFrame(data=myts_M_count, index=myts_M_count.index, columns=['profiles_posted'])
#print mydf_M_count.ix[:]

print type(mydf_M_count)
print type(mydf_M_avg)
#combine to create a monthly time series dataframe
monthly_df = pd.merge(pd.DataFrame(mydf_M_count),mydf_M_avg,right_index=True, left_index=True)
print monthly_df.ix[:]
```
