```python
import os
import re
import numpy as np
import pandas as pd
import urllib2
from datetime import datetime
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

from scipy import stats
from datetime import datetime
import statsmodels.formula.api as smf #allows R syntax for regressions
#print statsmodels.__version
```

```python
#folders within folder from which code is run
downloadfolder = "/Watsi_data_downloaded/"
processedfolder = "/Watsi_data_processed/"
resultsfolder = "/Summary_plots_and_tables/"
ppath = os.getcwd()

#file containing main data of interest
towrite = "Data_summary.csv" 
towrite_path = ppath + processedfolder + towrite
```

```python
# use this for floating figures:
#%matplotlib

# use this for inlined figures:
%matplotlib inline
```

```python
mydf = pd.read_csv(towrite_path, parse_dates=True)
mydf['date_posted'] = pd.to_datetime(mydf['date_posted'])
print mydf.ix[:5]
```

```python
#run a very simple regression to estimate effects of regressors on outcome
#import statsmodels.formula.api as smf

# Fit regression model (R syntax works)
results = smf.ols('dollars_per_day ~ \
                  day_posted + C(region) + maleness + treat_cost + patient_age + smile_scale', data=mydf).fit()

# Fit regression model (statsmodel syntax
#nona = mydf.dropna(axis=0)
#results = smf.OLS(nona.dollars_per_day, nona[['treat_cost','smile_scale']]).fit()

# Inspect the results
print results.summary()

# model after dropping insignificant terms (backwards selection process)
results = smf.ols('dollars_per_day ~ \
                  treat_cost + patient_age + smile_scale', data=mydf).fit()
print results.summary()
```

```python
mydf = pd.read_csv(towrite_path, parse_dates=True)
# create smile categories
bins = [-1, 0.2, 0.4, 0.6, 0.8, 1]
smile_cat_names = ["strongly_negative","negative","neutral","positive","strongly_positive"]
smile_dums = pd.get_dummies(pd.cut(mydf.smile_scale, bins, labels=smile_cat_names))
#print smile_dums.ix[:20]
mydf = pd.merge(mydf,smile_dums,left_index=True,right_index=True)
#print mydf.ix[:20]

results = smf.ols('dollars_per_day ~ \
                  treat_cost + patient_age + \
                  strongly_negative + negative + positive + strongly_positive', data=mydf).fit()
print results.summary() #strange/ difficult to explain results
```

```python
mydf = pd.read_csv(towrite_path, parse_dates=True)
# create smile categories
bins = [0, .45, .55, 1]
smile_cat_names = ["negative","neutral","positive"]
smile_dums = pd.get_dummies(pd.cut(mydf.smile_scale, bins, labels=smile_cat_names))
#print smile_dums.ix[:20]
mydf = pd.merge(mydf,smile_dums,left_index=True,right_index=True)
#print mydf.ix[:20]

results = smf.ols('dollars_per_day ~ \
                  treat_cost + patient_age + \
                  negative + positive', data=mydf).fit()
print results.summary() #being neutral is better than being positive, and maybe better than being negative
```

```python
mydf = pd.read_csv(towrite_path, parse_dates=True)
# create smile categories
bins = [0, .5, 1]
smile_cat_names = ["negative_or_neutral","positive"]
smile_dums = pd.get_dummies(pd.cut(mydf.smile_scale, bins, labels=smile_cat_names))
#print smile_dums.ix[:20]
mydf = pd.merge(mydf,smile_dums,left_index=True,right_index=True)
print mydf[['smile_scale','negative_or_neutral','positive']].ix[:20]

results = smf.ols('dollars_per_day ~ \
                  treat_cost + patient_age + \
                  positive ', data=mydf).fit()
print results.summary() #no effect (possible neg/neutral is better than being positive)
```
